```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import numpy as np
```

## Prepare the dataset

```python
boston = load_boston()
data = pd.DataFrame(boston.data)
columns = []
for feature in boston.feature_names:
    columns.append(feature.lower())
    
data.columns = columns    
data['price'] = pd.Series(boston.target)
```

## Calculate the baseline MSE using average price

```python
data['y_baseline'] = data.price.mean()
metrics.mean_squared_error(data.price, data.y_baseline)
```

## Begin building a model using Linear Regresion

```python
features = data.columns[0:-2]
target = data.columns[-2]
X = data[features]
# X = data[['rm', 'dis', 'tax', 'ptratio']]
y = data[target]
lr = LinearRegression()
lr.fit(X, y)
```

## Calculate the new MSE using the model to predict price 

```python
data['y_predict'] = lr.predict(X)
metrics.mean_squared_error(data.price, data.y_predict)
```

```python
output = pd.DataFrame
output = data[features]
output['price'] = data[target]
```
